# JSON

```python
var JSONUtils = {
    arrayToJson: function(array) {
        try {
            var jsonString = JSON.stringify(array);
            Watch.Log("数组已成功转换为JSON。", 3);
            return jsonString;
        } catch (e) {
            Watch.Log("转换数组为JSON时出错：" + e.message, 1);
            return null;
        }
    },

    sortJsonArray: function(jsonArray, field, ascending) {
        try {
            var sortedArray = jsonArray.sort(function(a, b) {
                var fieldA = a[field];
                var fieldB = b[field];
                if (fieldA < fieldB) {
                    return ascending ? -1 : 1;
                }
                if (fieldA > fieldB) {
                    return ascending ? 1 : -1;
                }
                return 0;
            });
            Watch.Log("JSON数组已根据字段 " + field + " 排序。", 3);
            return sortedArray;
        } catch (e) {
            Watch.Log("排序JSON数组时出错：" + e.message, 1);
            return null;
        }
    },

    splitJsonArrayToFile: function(jsonArray, field, filePathPrefix) {
        try {
            var groups = {};
            for (var i = 0; i < jsonArray.length; i++) {
                var item = jsonArray[i];
                var fieldValue = item[field];
                if (!groups[fieldValue]) {
                    groups[fieldValue] = [];
                }
                groups[fieldValue].push(item);
            }

            var fso = new ActiveXObject("Scripting.FileSystemObject");

            for (var key in groups) {
                if (groups.hasOwnProperty(key)) {
                    var jsonStr = JSON.stringify(groups[key], null, 2);
                    var filePath = filePathPrefix + "_" + key + ".json";
                    var file = fso.CreateTextFile(filePath, true);
                    file.Write(jsonStr);
                    file.Close();
                    Watch.Log("JSON文件已保存：" + filePath, 3);
                }
            }

            Watch.Log("JSON数组已根据字段 " + field + " 分割并保存为文件。", 3);
        } catch (e) {
            Watch.Log("根据字段分割并保存JSON数组时出错：" + e.message, 1);
        }
    }
};

// 使用示例
var array = [
    { "name": "Alice", "age": 30, "city": "New York" },
    { "name": "Bob", "age": 25, "city": "Los Angeles" },
    { "name": "Charlie", "age": 35, "city": "New York" },
    { "name": "David", "age": 30, "city": "Los Angeles" }
];

// 将数组转换为JSON
var jsonString = JSONUtils.arrayToJson(array);
Watch.Log("转换后的JSON字符串：" + jsonString, 3);

// 将JSON字符串解析为JSON对象
var jsonArray = JSON.parse(jsonString);

// 根据age字段进行升序排序
var sortedArray = JSONUtils.sortJsonArray(jsonArray, "age", true);
Watch.Log("排序后的JSON数组：" + JSON.stringify(sortedArray, null, 2), 3);

// 根据city字段分割并保存为文件
JSONUtils.splitJsonArrayToFile(jsonArray, "city", "C:\\path\\to\\your\\output\\json_city");
```
