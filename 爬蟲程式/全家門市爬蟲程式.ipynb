```python
def fetch_area_ids(driver, url):
    area_ids = []
    # 启动WebDriver
    driver.get(url)
    # 使用 XPath 查找所有 class 包含 "city" 的 div 元素
    cities = driver.find_elements(By.XPATH, "//div[contains(@class, 'city')]")

    # 遍历找到的元素并打印城市名称
    for city in cities:
        # 假设城市名称就是<a>标签内的文本
        city_name = city.text
        print(f"City Name: {city_name}")
        area_ids.append(city_name)
    return area_ids

url = 'https://www.family.com.tw/Marketing/storemap/?v=1'
driver = webdriver.Chrome()
fetch_area_ids(driver,url)
```

```python
def fetch_table_data_with_pagination(driver,url):
    driver.get(url)
    time.sleep(5)
    district_table_data = []
    try:
        for key,val in area_district.items():
            for dis in val:
                driver.execute_script("searchByCity('"+key+"','"+dis+"')")
                time.sleep(1)
                soup = BeautifulSoup(driver.page_source, "html.parser")
                page_bu = soup.find("div", id="page_bu_content")
                page_count = len(page_bu.find_all("li")) if page_bu else 0
                page = 1
                while page <= page_count:
                    soup = BeautifulSoup(driver.page_source, "html.parser")
                    tables = soup.find_all("table", class_="graybigbox")
                    trs = tables[0].find_all("tr")[1:]
                    # 使用步长为2的切片来遍历 trs 列表
                    for i in range(0, len(trs), 2):
                        # 获取当前和下一个 <tr> 元素
                        tr_info = trs[i]
                        # tr_shop = trs[i + 1] if i + 1 < len(trs) else None
                        
                        shop_name = tr_info.find_all("td")[0].text.strip()
                        shop_info_text = tr_info.find_all("td")[1].text.strip()
                        
                        # 使用正则表达式匹配店铺号、服务编号、地址和电话
                        shop_number = re.search(r"店舖號：(\d+)", shop_info_text)
                        service_number = re.search(r"服務編號：(\d+)", shop_info_text)
                        address = re.search(r"地址：(.*?) 電話：", shop_info_text)
                        phone = re.search(r"電話：([\d ,\-]+)", shop_info_text)
                        
                        # 获取匹配的结果，如果没有找到则返回None
                        shop_number = shop_number.group(1) if shop_number else "未提供"
                        service_number = service_number.group(1) if service_number else "未提供"
                        address = address.group(1).strip() if address else "未提供"
                        phone = phone.group(1).strip() if phone else "未提供"
                                        
                        
                        
                        # 找到包含服务代码的 td 元素
                        services_td = trs[i].find_all("td")[4]
                        # 获取所有服务代码
                        services_spans = services_td.find_all("span")
                        
                        services = []
                        for span in services_spans:
                            service_code = span.get('class', [None])[0]  # 获取 span 的 class
                            if service_code and service_code in service_mapping:
                                services.append(service_mapping[service_code])
                            else:
                                services.append("未知服务")  # 对于未在映射表中找到的类名

                        # 将提取的信息组装成一个字典，以便于后续处理
                        shop_data = {
                            "shop_name": shop_name,
                            "shop_number": shop_number,
                            "address": address,
                            "phone": phone,
                            "services": ", ".join(services)
                        }
                        print(shop_data)
                        district_table_data.append(shop_data)
                    soup = ''
                    try:
                        if page == page_count:
                            break
                        else:
                            page+=1
                            driver.execute_script('nextPage()')
                    except Exception as e:
                        if e == 'Alert Text: 已經是最後一頁':
                            break  # 没有找到下一页时退出循环
    except Exception as e:
        print(f"Error fetching table data for district {dis}: {e}")
    return district_table_data

url = 'https://www.family.com.tw/Marketing/storemap/?v=1'
driver = webdriver.Chrome()

fetch_table_data_with_pagination(driver,url)
```

```python
import csv
import json

def json_to_csv(json_data_path, csv_file_path):
    """
    从 JSON 文件读取数据，将其转换为 CSV 文件。
    
    :param json_data_path: JSON 文件的路径。
    :param csv_file_path: 输出 CSV 文件的路径。
    """
    # 从 JSON 文件读取数据
    with open(json_data_path, 'r', encoding='utf-8') as json_file:
        json_data = json.load(json_file)

    # 确保 json_data 是列表
    if not isinstance(json_data, list):
        raise ValueError("JSON 文件内容应该是一个字典列表")
    
    # 检查是否有数据来处理
    if not json_data:
        raise ValueError("JSON 数据为空")

    # 打开 CSV 文件准备写入
    with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
        # 如果 JSON 数据非空，则创建 CSV 写入器
        writer = csv.DictWriter(csv_file, fieldnames=json_data[0].keys())
        
        # 写入列标题
        writer.writeheader()
        
        # 写入 JSON 数据
        for row in json_data:
            writer.writerow(row)

# 示例用法
json_data_path = ".json"  # 假设你的 JSON 文件名为 input.json
csv_file_path = ".csv"
json_to_csv(json_data_path, csv_file_path)
```
